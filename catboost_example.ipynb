```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

import catboost
import time as tm
```

```python
train = pd.read_csv(
    '/data/temp/train.csv',
    )
```

```python
train
```

```python
cat_features = []

for c in train.columns:
    if train[c].dtype == "object":
        train[c] = train[c].astype("category")
        cat_features.append(c)

X = train.drop(columns = ['survived'])
y = train['survived'].to_numpy()
```

```python
cat_features
```

```python
estimator = catboost.CatBoostClassifier(
    cat_features = cat_features,
    nan_mode = 'Min',
    task_type = 'GPU', #using GPU
    devices = '0', # ID of used GPU
    )
```

https://catboost.ai/en/docs/concepts/python-reference_catboostclassifier#cat_features

```python
param_grid = {
    'learning_rate': [0.1,0.01],
    'num_trees': [50, 100],
}
```

```python
gbm = GridSearchCV(
    estimator, 
    param_grid, 
    cv = 5,
    scoring = 'roc_auc')
```

```python
start_time = tm.time()

gbm.fit(X, y)

running_time = tm.time() - start_time
```

```python
print('running time: %f seconds'%(running_time))
```

```python
gbm.best_score_
```

```python
gbm_best = gbm.best_estimator_
```

```python
features = zip(
    gbm_best.feature_names_,
    gbm_best.feature_importances_,
)
```

```python
for n, i in features:
    print(n, '\t', i)
```
